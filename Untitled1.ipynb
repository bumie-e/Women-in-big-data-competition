```python
def fill_outliers(train, *tuple_val):
    num = len(tuple_val)
    train = train
    if tuple_val[1] == 0:
        train[train < tuple_val[0]] = tuple_val[0]
    elif tuple_val[0] == 0:
        train[train > tuple_val[1]] = tuple_val[1]
    else:
        train[train < tuple_val[0]] = tuple_val[0]
        train[train > tuple_val[1]] = tuple_val[1]
    
    return train
```
